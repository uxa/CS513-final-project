```python
map_months = {
    'January': '01/01/{}',
    'February': '02/01/{}',
    'March': '03/01/{}',
    'April': '04/01/{}',
    'May': '05/01/{}',
    'June': '06/01/{}',
    'July': '07/01/{}',
    'August': '08/01/{}',
    'September': '09/01/{}',
    'October': '10/01/{}',
    'November': '11/01/{}',
    'December': '12/01/{}'
}
```

```python
temp = list()
```

```python
all_colms = ['Season1 Start Date', 'Season1 End Date', 'Season2 Start Date', 'Season2 End Date', 'Season3 Start Date', 'Season3 End Date', 'Season4 Start Date', 'Season4 End Date']
for coln in all_colms:
    temp = list()
    for index, value in df.iterrows():
        col = value[coln]
        col_update_time = value['Update date']
        if col in map_months:
            temp.append(map_months.get(col).format(col_update_time.split('-')[0]))
        else:
            temp.append(col)
    df[coln] = temp
```

```python
cols_to_drop = ['Season1Time', 'Season2Time', 'Season3Time', 'Season4Time']
for col in cols_to_drop:
    df.drop(col, axis = 1, inplace = True)
```

```python
df.to_csv('Farmers-Market-Final.csv')
```
